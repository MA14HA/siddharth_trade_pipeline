```python
import pandas as pd

# IMPORTANT: Replace 'import_data_2017_2025.csv' with the actual name of your raw file.
file_path = "../data/raw/sample_data.csv"

# Assuming CSV format. Use pd.read_excel() if your file is .xlsx.
df = pd.read_csv(file_path)

print(f"Dataframe loaded with {len(df)} rows and {len(df.columns)} columns.")
```

```python
# 1. Preview the first 5 rows
print("--- Data Head ---")
print(df.head())

# 2. Check data types and identify missing values (Nulls)
print("\n--- Data Info (Types & Nulls) ---")
df.info()

# 3. Check summary statistics for numerical consistency (outliers, negatives)
print("\n--- Numerical Summary ---")
print(df.describe())
```

```python
print(df.columns.tolist())
```

```python
DATE_COL = 'DATE'
print("\n--- DATE Examples ---")
print(df[DATE_COL].sample(5).tolist())
# Note the exact format (e.g., DD/MM/YYYY)
```

```python
DESCRIPTION_COL = 'GOODS DESCRIPTION'

print("\n--- GOODS DESCRIPTION Patterns (20 Samples) ---")
# Use random_state for reproducible sampling
for desc in df[DESCRIPTION_COL].sample(20, random_state=42).tolist():
    print(f"- {desc}")
```
